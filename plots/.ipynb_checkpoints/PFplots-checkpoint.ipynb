```python
import ast
import matplotlib as mpl
from matplotlib import pyplot as plt
import matplotlib.patches as patches
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
gt = pd.read_csv(open("vehicle.out",'rb'), header=None, delim_whitespace=True)
```

```python
lm = pd.read_csv(open("../data/map_data.txt",'rb'), header=None, delim_whitespace=True)
```

```python
all_particles = []
best_particles = []
with open("pf50.out", "r") as f:
    while True:
        line1 = f.readline()
        line2 = f.readline()
        if line1 == "" or line2 == "":
            break
        all_particles.append(ast.literal_eval(line1))
        best_particles.append(ast.literal_eval(line2))

best_particles = np.array(best_particles)
all_particles = np.array(all_particles)
```

```python
color_cycle = mpl.rcParams['axes.prop_cycle']()
blue = next(color_cycle)['color']
orange = next(color_cycle)['color']
```

```python
gt_color = 'black'
gt_label = "Ground Truth Vehicle Path"
pf_label = "Best Particle"

x_label = "x"
y_label = "y"

arrow_scale = 6
plot_every_timestep = 20

f, ax1 = plt.subplots(1, 1, figsize=(8, 5), dpi=300)
f.tight_layout()

ax1.plot(gt.iloc[:-plot_every_timestep//2,0], gt.iloc[:-plot_every_timestep//2,1], c=gt_color, lw=2, label=gt_label)
ax1.scatter(lm.iloc[:,0], lm.iloc[:,1], zorder=2, facecolors='none', edgecolors=blue, s=16, label="Landmarks")

for i in range(0, len(best_particles), plot_every_timestep):
    x = best_particles[i,0]
    y = best_particles[i,1]
    dx = np.cos(best_particles[i,2]) * arrow_scale
    dy = np.sin(best_particles[i,2]) * arrow_scale
    if i == 0:
        l = pf_label
    else:
        l = None
    ax1.add_patch(patches.Arrow(x - dx, y - dy, 2 * dx, 2 * dy, width=12, edgecolor="white", facecolor=orange, zorder=3, label=l))

ax1.set_xlabel(x_label)
ax1.set_ylabel(y_label)
ax1.set_title("Particle Filter for Vehicle Localization")
ax1.legend()

ax1.axis('equal')

plt.show()
```
